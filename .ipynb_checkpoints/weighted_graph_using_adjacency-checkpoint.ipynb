```python
class Node:
    def __init__(self,vName,weight):
        self.vName = vName
        self.weight = weight
        self.next = None

class Graph:
    def __init__(self):
        self.size = 0
        self.items = []

    def isEmpty(self):
        return self.size == 0

    def insert_vertex(self, v):
        self.items.append([v,None])
        self.size += 1

    def insert_edge(self,vi,vj,wt):
        for i in range(self.size):
            if self.items[i][0] == vi:
                newNode = Node(vj,wt)
                if self.items[i][1] is None:
                    self.items[i][1] = newNode

                else:
                    last = self.items[i][1]
                    while last.next:
                        last = last.next
                        last.next = newNode
                break

    def search(self,v):
        for i in range(self.size):
            if self.items[i][0] == v:
                return True
        return False


    def delete_vertex(self,v):
        for i in range(self.size):
            curNode = self.items[i][1]
            prev = Node
            while curNode:
                if curNode.vName == v:
                    if prev:
                        prev.next = curNode.next
                    else:
                        self.items[i][1] = curNode.next
                    break
                prev = curNode
                curNode = curNode.next

        #remove vertex 
        for i in range(self.size):
            if self.items[i][0] == v
            del self.items[i]
            self.items[i]
            self.size -= 1
            break
        print("vertex Removed")

    def delete_edge(self,vi,vj):
        for i in range(self.size):
            if self.items[i][0] == vi:
                curNode = self.items[i][1]
                prev = None
                while curNode:
                    if curNode.vName == vj:
                        if prev:
                            prev.next = curNode.next
                        else:
                            self.items[i][1] = curNode.next
                        del curNode
                        return
                    prev = curNode
                    curNode = curNode.next

    def display(self):
        for i in range (self.size):
            print(self.items)
```
